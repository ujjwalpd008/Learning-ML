# Box-Plot with Whiskers

Its another way of visualizing 1D scatter plot<br>

Whiskers-> its nothing but the box with lines representing Quantiles(0,25,50,75,100).
<br>

In Histograms its some what hard to identify percentiles in that graphs, with this its easy to identify percentiles.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv');
```

```python
sns.boxplot(x='species',y='petal_length',data=iris)
plt.show();
```

In the above Graph all the percentiles/quantiles are highlighted
<br>

Observation: Through this we can easily identify what percentile of Flowers with petal length x will fall under the given interval

# Violin Plots

Its a combination of Histogram_pdf and Box plots

```python
sns.violinplot(x='species',y='petal_length',data=iris,size=10)
plt.show();
```

In the above image we can observe Both Box plot and Pdf/Histogram(bell curve)plot
