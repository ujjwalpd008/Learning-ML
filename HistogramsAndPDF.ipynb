## Histogram and PDF(Probability Density Function) EDA

#### We have checked 2d, 3d, pairplots, Lets check 1D plot using histogram

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv');
```

```python
#size is size of the graph

#sns.distplot (Distribution plot) in below fn is y axis i.e the count of points between intervals of x
#Example: if there are 4 flowers with PL range 1 - 1.5 the y axis will 
# be 4 for all the points between 1 - 1.5
    
sns.FacetGrid(iris,hue='species',size=6) \
    .map(sns.distplot,"petal_length") \
    .add_legend();
plt.show();
```

### In the above fig the smooth line is PDF(probability density fn)

Obesevation: As we considered only one feature its hard to distinguish Versicolor and Virginica
